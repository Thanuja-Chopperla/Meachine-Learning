```python
# QUESTION 7

!pip install hmmlearn

import numpy as np
from hmmlearn import hmm

# Hidden states (Rainy or Sunny)
states = ["Rainy", "Sunny"]
n_states = len(states)

# Observable activities
observations = ["Walk", "Shop", "Clean"]
n_observations = len(observations)

# Transition probabilities
transition_prob = np.array([
    [0.7, 0.3],
    [0.4, 0.6]
])

# Emission probabilities
emission_prob = np.array([
    [0.1, 0.4, 0.5],  # Rainy emits Walk/Shop/Clean
    [0.6, 0.3, 0.1]   # Sunny emits Walk/Shop/Clean
])

# Start probabilities
start_prob = np.array([0.6, 0.4])

# Create HMM model
model = hmm.MultinomialHMM(n_components=n_states, n_iter=100, tol=0.01)
model.startprob_ = start_prob
model.transmat_ = transition_prob
model.emissionprob_ = emission_prob
model.n_trials = 1  # Needed for new version

# âœ… FIX: Represent observations as one-hot encoded vectors
# Walk = [1, 0, 0], Shop = [0, 1, 0], Clean = [0, 0, 1]
obs_sequence = np.array([
    [1, 0, 0],  # Walk
    [0, 1, 0],  # Shop
    [0, 0, 1]   # Clean
])

# Decode sequence using Viterbi algorithm
logprob, hidden_states = model.decode(obs_sequence, algorithm="viterbi")

print("Observed sequence:")
print(["Walk", "Shop", "Clean"])
print("\nMost likely hidden states (weather):")
print([states[i] for i in hidden_states])
print("\nLog probability of this sequence:", logprob)

# Generate a random sequence of 5 steps
X, Z = model.sample(5)
print("\nGenerated observation sequence:")
for x in X:
    obs_index = np.argmax(x)
    print(observations[obs_index], end="  ")

print("\nGenerated hidden states:")
print([states[i] for i in Z])
```
