```python
#QUESTION 5
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import scipy.cluster.hierarchy as sch

# Generate synthetic data (3 clusters)
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Visualize the data
plt.scatter(X[:, 0], X[:, 1], s=40)
plt.title("Generated Data")
plt.show()
```

```python
# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=0)
labels_kmeans = kmeans.fit_predict(X)

# Visualize clusters
plt.scatter(X[:, 0], X[:, 1], c=labels_kmeans, s=40, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

```python
# Apply Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=0)
labels_gmm = gmm.fit_predict(X)

# Visualize clusters
plt.scatter(X[:, 0], X[:, 1], c=labels_gmm, s=40, cmap='plasma')
plt.title("Gaussian Mixture Model Clustering")
plt.show()
```

```python
# Apply Agglomerative Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels_hier = hierarchical.fit_predict(X)

# Visualize clusters
plt.scatter(X[:, 0], X[:, 1], c=labels_hier, s=40, cmap='cool')
plt.title("Hierarchical Clustering")
plt.show()

# Dendrogram (optional visualization)
plt.figure(figsize=(10, 5))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score

print("K-Means ARI:", adjusted_rand_score(y_true, labels_kmeans))
print("GMM ARI:", adjusted_rand_score(y_true, labels_gmm))
print("Hierarchical ARI:", adjusted_rand_score(y_true, labels_hier))
```
